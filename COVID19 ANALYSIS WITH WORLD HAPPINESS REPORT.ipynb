#### Considering only France

```python
### Let's take France and try to understand better

covid_data_agg.loc["France"].plot()
```

```python
plt.figure(figsize=(10,8))
plt.title("Plot for Number of cases found in very first 100 days")
plt.ylabel("Nuber of Cases")
plt.xlabel("Dates")
covid_data_agg.loc["France"][:100].plot(figsize=(8,4))
```

### Plotting the first derivative of the curve
#### Let us find the maximum rate at which covid hit the coutry/region



```python
covid_data_agg.loc["France"].diff().plot(figsize=(8,4))
plt.legend()
plt.ylabel("Nuber of Cases")
plt.xlabel("Dates")
```

<b>Maximum number of infected cases in One day</b>

```python
print("France - ",covid_data_agg.loc["France"].diff().max())
```

#### States that more than 26849 cases hit in France

```python
### Let us take some countries for sample

print("China -",covid_data_agg.loc["China"].diff().max())
print("Italy -",covid_data_agg.loc["Italy"].diff().max())
print("UK -",covid_data_agg.loc["United Kingdom"].diff().max())
print("US -",covid_data_agg.loc["US"].diff().max())
```

## Maximum Rate of Case for all countries

```python

countries = list(covid_data_agg.index)
max_infections = []
for i in countries:
    max_infections.append(covid_data_agg.loc[i].diff().max())
covid_data_agg["max_infections"] = max_infections 
covid_data_agg.head()
```

#### Creating a column called "max_infections" which shows the maximum number of infections per day in each country

<br>--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
### Considering only the necessary "max_infections" column

```python
corona_infection = pd.DataFrame(covid_data_agg["max_infections"])
corona_infection.head()
```

### Task VII
### Merging the datasets
<p>Setting the index as "Country/Region" in order to join both the datasets based on that column</b>

```python
happ_report.rename(columns = {"Country or region":"Country/Region"},inplace = True)
```

```python
happ_report.set_index("Country/Region",inplace=True)
```

```python
print(corona_infection.shape)
print(happ_report.shape)
```

<b>Note : <br>Since the shape is dissimilar join by using inner join

```python
df = corona_infection.join(happ_report,how="inner")
df.head()
```

### Correlation of dataset

```python
df.corr()
```

```python
## For better understanding - visualizing the correlation
f,ax = plt.subplots(figsize = (8,5))
sns.heatmap(df.corr(),annot = True,fmt='.2f',linewidths=.3,ax=ax)
```

<b>GDP per capita vs max_infections</b>

```python
x=df["GDP per capita"]
y=df["max_infections"]
sns.regplot(x,np.log(y),data=df)
```

<b>Social support vs max_infections

```python
x=df["Social support"]
y=df["max_infections"]
sns.regplot(x,np.log(y))
```

<b>Freedom to make life choices vs max_infections

```python
x=df["Freedom to make life choices"]
y=df["max_infections"]
sns.regplot(x,np.log(y))
```

```python
x=df["Overall rank"]
y=df["max_infections"]
sns.regplot(x,np.log(y))
```

### Result

- It is found that the <b>OVERALL RANK and MAXIMUM INFECTIONS ARE INVERSELY PROPORTIONAL</b>
   - (i.e) - <b>maximum number</b> of COVID-19 infections are in the <b>most DEVELOPED COUNTRIES compared to LESS DEVELOPED and DEVELOPING COUNTRIES.</b>


- It may be <b>due to the factors such as 
    - less people being tested in the developing countries in comparison to the developed countries
    - lack of kits available in developing countries.</b>
    
 - This can be verified by using the cumulative death rate of each countries
